```python
import random
import os
# Split yolo v5 data into train, valid and test 
# data_path: path to the data folder
# return: train_path, valid_path, test_path
data_path = 'data/'
out_dir = "yolov5/"
train_path = out_dir + 'DATASET V2'
if not os.path.exists(train_path):
    os.mkdir(train_path)

valid_path = out_dir + 'valid'
if not os.path.exists(valid_path):
    os.mkdir(valid_path)
    
test_path = out_dir + 'test'
if not os.path.exists(test_path):
    os.mkdir(test_path)

# Split and copy images and labels
all_files = os.listdir(data_path)
files_base = [a[:-4] for a in all_files if a[-3:] == 'jpg']
# Shuffle images
random.shuffle(files_base)
# Split images into train, valid and test
train_images = files_base[:int(len(files_base) * 0.94)] 
valid_images = files_base[int(len(files_base) * 0.94):int(len(files_base) * 0.97)]
test_images = files_base[int(len(files_base) * 0.97):]
```

```python
import shutil
from IPython.core.debugger import set_trace
import os 
# Copy files to train, valid and test folders
for image in train_images:
#     set_trace()
    if os.path.exists(data_path+'DATASET_V2/' + image+".txt"):
        shutil.copy(data_path+'DATASET_V2/' + image+".jpg", train_path)
        shutil.copy(data_path+'DATASET_V2/' + image+".txt", train_path)
    
for image in valid_images:
    if os.path.exists(data_path+'DATASET_V2/' + image+".txt"):
        shutil.copy(data_path+'DATASET_V2/' + image+".jpg", valid_path)
        shutil.copy(data_path+'DATASET_V2/' + image+".txt", valid_path)
    
for image in test_images:
    if os.path.exists(data_path+'DATASET_V2/' + image+".txt"):
        shutil.copy(data_path+'DATASET_V2/' + image+".jpg", test_path)
        shutil.copy(data_path+'DATASET_V2/' + image+".txt", test_path)
```

```python
f = open('data.yaml', 'w')
f.write("train: train\nval: valid\nnc: 16")
f.close()
```
